# Importing modules

```python
import numpy as np
import sep
import fitsio # need to use astropy fits routine instead
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline

rcParams['figure.figsize'] = [10., 8.]
```

# Reading and displaying FITS Image

```python
# reading in the test image
fits_file = fits.util.get_testdata_filepath('test0.fits') # need a test file

data = fitsio.read("image.fits")

# hdul.info()

# displaying the test image
m, s = np.mean(data), np.std(data)
plt.imshow(data, interpolation='nearest', cmap='gray', vmin=m-s, vmax=m+s, origin='lower')
plt.colorbar()
```

# Background Subtraction

```python
# measure the background (varying) of the image
bkg = sep.Background(data)
print(bkg.globalback)
print(bkg.globalrms)
```

```python
bkg_image = bkg.back()
plt.imshow(bkg_image, interpolation='nearest', cmap='gray', origin='lower')
```

```python
# calculase background noise as a 2d array
bkg_rms = bkg.rms()
plt.imshow(bkg_rms, interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar()
```

```python
data_sub = data - bkg
```

# Detecting Objects

```python
objects = sep.extract(data_sub, 1.5, err=bkg.globalrms)
len(objects)
```

```python
from matplotlib.patches import Ellipse

# plot background subtracted image
fig, ax = plt.subplots()
m, s = np.mean(data_sub), np.std(data_sub)
im = ax.imshow(data_sub, interpolation='nearest', cmap='gray',
               vmin=m-s, vmax=m+s, origin='lower')

for i in range(len(objects)):
    e = Ellipse(xy=(objects['x'][i], objects['y'][i]),
                width=6*objects['a'][i],
                height=6*objects['b'][i],
                angle=objects['theta'][i] * 180 / np.pi)
    e.set_facecolor('none')
    e.set_edgecolor('red')
    ax.add_artist(e)
```
